## We can create new column and remove less relevant columns

```python
df = df_raw.copy()
```

```python
print(df.columns)
```

```python
df.drop(['Date', 'Time', 'NMHC(GT)'],axis= 1, inplace= True)
```

## Let's replace Missing values

Looking into the dataset, we can identify that the missing values are stored as -200. Later we can check the distribution of values in the columns and decided the suitable values to replace from mean & median (No logic in considering mode, for this dataset)

```python
df.replace(-200,np.NaN, inplace= True)
```

```python
print(df.isna().sum())
```

```python
#Ground_Truth Columns
GT = ['CO(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)']

#Sensor Columns
Sensors = ['PT08.S1(CO)', 'PT08.S2(NMHC)', 'PT08.S3(NOx)', 'PT08.S4(NO2)', 'PT08.S5(O3)']

#Environmental columns

EVC = ['T', 'RH', 'AH']
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for idx, col in enumerate(GT):
    sns.histplot(df[col], kde=True, ax=axes[idx])
    axes[idx].set_title(col)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for idx, col in enumerate(Sensors):
    sns.histplot(df[col], kde=True, ax=axes[idx])
    axes[idx].set_title(col)

# Remove unused subplot (since 5 sensors, 6 slots)
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, col in enumerate(EVC):
    sns.histplot(df[col], kde=True, ax=axes[idx])
    axes[idx].set_title(col)

plt.tight_layout()
plt.show()
```

# Replacing NaN values

We with Mean for Noral distribution/bell shaped curves, and median for skewed curves

```python
for col in Sensors:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
for col in EVC:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
for col in GT:
    df[col].fillna(df[col].median(), inplace=True)
```
